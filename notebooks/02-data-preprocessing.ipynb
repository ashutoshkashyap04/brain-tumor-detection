```python
from pathlib import Path

processed_dir = Path("../data/processed")

train_processed = processed_dir / "train"
val_processed = processed_dir / "validation"
test_processed = processed_dir / "test"

for folder in [train_processed, val_processed, test_processed]:
    folder.mkdir(parents=True, exist_ok=True)
```

```python
classes = ["glioma_tumor", "meningioma_tumor", "no_tumor", "pituitary_tumor"]

for dataset_folder in [train_processed, val_processed, test_processed]:
    for class_name in classes:
        (dataset_folder / class_name).mkdir(parents=True, exist_ok=True)
```

```python
import shutil
from sklearn.model_selection import train_test_split
```

```python
from pathlib import Path

raw_train_dir = Path("../data/raw/Training")

for class_name in classes:

    # All images of the current class
    images = list((raw_train_dir / class_name).glob("*"))

    # Split into train and validation
    train_images, val_images = train_test_split(
        images,
        test_size=0.2,
        random_state=42,
        shuffle=True
    )

    # Copy training images
    for img in train_images:
        shutil.copy(img, train_processed / class_name / img.name)

    # Copy validation images
    for img in val_images:
        shutil.copy(img, val_processed / class_name / img.name)

print("Train and Validation images copied successfully.")
```

```python

raw_test_dir = Path("../data/raw/Testing")

for class_name in classes:

    images = list((raw_test_dir / class_name).glob("*"))

    for img in images:
        shutil.copy(img, test_processed / class_name / img.name)

print("Testing images copied successfully.")
```
