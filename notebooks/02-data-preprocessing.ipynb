```python
# defining fixed image size

IMG_HEIGHT = 224
IMG_WIDTH = 224

IMG_SIZE = (IMG_HEIGHT, IMG_WIDTH)
```

```python
# defining the paths

train_dir = "../data/raw/Training"
test_dir = "../data/raw/Testing"
```

```python
# load the training dataset

import tensorflow as tf

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="training",
    seed=42,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)
```

```python
# creating validation dataset

val_dataset = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="validation",
    seed=42,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)
```

```python
# load the test dataset

test_dataset = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)
```

```python
# Checking the detected classes
class_names = train_dataset.class_names
print(class_names)
```

```python
# Verifying the shape of one batch

for images, labels in train_dataset.take(1):    #train_dataset.take(1) creates a new dataset containing only the first batch.   
    print(images.shape)
    print(labels.shape)
```
